```python
import matplotlib.pyplot as plt
import keras
import tensorflow as tf
import numpy as np
%matplotlib inline
```

```python
#데이터 셋 불러오기
(x_train, y_train), (x_test, y_test) = keras.datasets.boston_housing.load_data()
```

```python
#데이터의 모양 확인

#히스토그램 그리기
#bins : 표시할 막대의 수
plt.hist(y_train, bins=20)
plt.xlabel('house price(1000$)')
plt.ylabel('number of houses')
plt.show()

#분포도 그리기
plt.plot(x_train[:,5], y_train, 'o')
plt.xlabel('# of rooms')
plt.ylabel('house price(1000$)')
```

```python
'''
우리가 다루는 데이터는 모두 범위가 제각각이다
같은 현상을 다루는 데이터 일지라도 단위가 달라지면 범위가 달라진다
이렇게 데이터의 축이 일치가 되지 않으면 좋은 성과를 내기는 어렵다

따라서 모든 데이터의 번위가 같아지도록 하는것을 
정규화(normalization)한다고 한고
평균과 얼마나 떨어져 있는지를 나타내는 것을
표쥰화(standardization)한다고 한다
'''

#데이터의 표준화
#axis 방향이 0 이면 행방향
x_train_mean = x_train.mean(axis=0)
x_train_std = x_train.std(axis=0)
y_train_mean = y_train.mean()
y_train_std = y_train.std()

x_train = (x_train - x_train_mean)/x_train_std
y_train = (y_train - y_train_mean)/y_train_std

x_test = (x_test - x_train_mean)/x_train_std
y_test = (y_test - y_train_mean)/y_train_std

#표준화 후의 데이터의 모양
plt.plot(x_train[:,5], y_train, 'o')
plt.xlabel('number of rooms')
plt.ylabel('house price(after standardization)')
```

```python
#모델 정의
#x와y는 우리의 데이터가 들어갈 자리
#w는 입력값과 곱해질 가중치 벡터
#이때, 예측 결과가 하나의 값이므로 벡터의 모양은 (13,1)
#matmul은 행렬의 곱셈을 나타냄

x = tf.placeholder(tf.float32, (None, 13), name='x')
y = tf.placeholder(tf.float32, (None, 1), name='y')
w = tf.Variable(tf.random_normal(shape=(13,1)))
pred = tf.matmul(x,w)

#오차 정의
#이때 오차는 평균제곱 오차로 실제값과 예측값의 차이를 제곱한 오차
#최적화할때 미분이 편해짐
loss = tf.reduce_mean((y-pred)**2)
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1)
train_step = optimizer.minimize(loss)
```

```python
#SGD를 실행하기 위해 batch 크기만큼 데이터를 분할
#호출될 때마다 batch크기만큼 데이터를 내보내기 위한 generator를 만듬

def get_batches(x, y, batch_size):
    n_data = len(x)
    
    #indices: 데이터의 순서를 무작위로 섞기위한 배열
    #arange는 등차수열을 만들어주는 함수
    #입력받은 x와 y를 indices로 파싱하면
    #해당 indices 순서로 배열이 섞임
    #ex) x = np.array([[1],[2],[3]])
    #    indices = np.array([1,2,0])
    #    x[indices] => np.array([[2],[3],[1]])
    indices = np.arange(n_data)
    np.random.shuffle(indices)
    x_shuffled = x[indices]
    y_shuffled = y[indices]
    
    for i in range(0, n_data, batch_size):
        #batch 크기만큼 분할
        x_batch = x_shuffled[i:i+batch_size]
        y_batch = y_shuffled[i:i+batch_size]
        yield x_batch, y_batch
```

```python
BATCH_SIZE = 32
step = 0
losses = []
steps = [i for i in range(100)]

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for epoch in range(100):
        #100번 반복
        for x_batch, y_batch in get_batches(x_train, y_train, 32):
            #iterator x_batch와 y_batch가 호출되면서 yiled된 반환값을 받음
            train_loss, _ = sess.run(
                [loss, train_step],
                feed_dict={
                    x: x_batch,
                    y: y_batch.reshape((-1, 1))
                }
            )
            step+=1
        losses.append(train_loss)
        pred_ = sess.run(
        pred,
        feed_dict={
            x: x_test
        }
    )
    print(train_loss)
```

```python
plt.rcParams['font.family'] = ['IPAexGothic']
plt.rcParams['font.size'] = 10*3
plt.rcParams['figure.figsize'] = [18, 12]

plt.plot(x_test[:,5], pred_, 'o', label='prediction')
plt.plot(x_test[:,5], y_test, 'x', label='real value')
plt.legend(loc='upper_left')
plt.xlabel('number of rooms')
plt.ylabel('house price')
```
